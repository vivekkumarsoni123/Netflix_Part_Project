```python
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('GOOGL_API_KEY')
model = genai.GenerativeModel('gemini-pro')
```

```python
import pathlib
import textwrap
from gtts import gTTS

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
from IPython.display import Audio, display


def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
#Step 1
def ask_question(name):
  ques = "Hey " + name + " what do you want ?"
  ques = input(ques)
  return ques;
```

```python
def classify_questions(ques):
  goahead_with_web_search = False

  device_list = ['alarm','reminder','message','call']
  personal_list = ['who are you','who created you']

  device = False
  for i in device_list:
    if i in ques:
      device = True

  personal_ques = False
  for i in personal_list:
    if i in ques:
      personal_ques = True


  if personal_ques:
    print("Hey I am Jarvis, created by Vivek")

  if device:
    print("This qestion is related to device things, which is not supported in our google assistance")
  else:
    goahead_with_web_search = True

  return goahead_with_web_search
```

```python
classify_questions("who are you")
```

```python
def ask_gemini(ques):
  modified_promp = 'Hey give me answer to this question ' + ques + 'in maximum of 40 words'
  response = model.generate_content(modified_promp)
  to_markdown(response.text)
  return response.text
```

```python
def speak(answer):
  tts = gTTS(answer)
  tts.save('audio.mp3')
  display(Audio('audio.mp3', autoplay=True))
```

```python
have_any_other_ques = 'y'
name = ''

while have_any_other_ques == 'y':

  if name == '':
    name = input("Hey what is your name")

  q = ask_question(name)

  go_ahead = classify_questions(q)
  answer = ''
  if go_ahead == True:
    answer = ask_gemini(q)
    speak(answer)
    print(answer)


  have_any_other_ques = input("Do you have any other question?")


```
